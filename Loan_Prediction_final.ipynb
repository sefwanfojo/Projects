```python
#Feature Engineering
```

```python
df['Years in current job'].unique()
```

```python
d_years={'10+ years':12, '4 years':4, '6 years':6, '5 years':5, '0':0, '3 years':3,'2 years':2, '< 1 year':1, '1 year':1, '7 years':7, '9 years':9, '8 years':8}
```

```python
df['Years in current job']=df['Years in current job'].map(d_years)
```

```python
df['Loan Status'].unique()
```

```python
d_lstatus={'Fully Paid':1,'Charged Off':0}
```

```python
df['Loan Status']=df['Loan Status'].map(d_lstatus)
```

```python
df["Purpose"].unique()
```

```python
df["Purpose"]=df["Purpose"].replace("other","Other")
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=df["Purpose"])
```

```python
df.head()
```

```python
df["Monthly Debt"]=df["Monthly Debt"].str.replace("$","")
df["Monthly Debt"]=df["Monthly Debt"].str.replace(",","")
```

```python
df["Monthly Debt"]=df["Monthly Debt"].astype(float)
```

```python
df.sample(10)
```

```python
x=df.drop(['Loan Status'],axis=1)
y=df["Loan Status"]
```

```python
df.isnull().sum()
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB
```

```python
x_train.fillna(x_train.median(), inplace=True)
```

```python
g=GaussianNB()
b=BernoulliNB()
```

```python
g.fit(x_train,y_train)
b.fit(x_train,y_train)
```

```python
predg=g.predict(x_test)
predb=b.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(predb,y_test)
```

```python
accuracy_score(predg,y_test)
```

```python
confusion_matrix(predg,y_test)
```

```python
confusion_matrix(predb,y_test)
```
